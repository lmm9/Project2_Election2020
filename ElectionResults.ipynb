```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np 
```

```python
#Importing election data 
election_data = "president_county_candidate(11212020).csv"
# Read our Data file with the pandas library
election_data_df = pd.read_csv(election_data, encoding="utf-8")
# Show just the header
election_data_df.head()
```

```python
# Read our Data file with the pandas library
election_data_df = pd.read_csv(election_data, encoding="utf-8")
# Show just the header
election_data_df.head()
```

```python
election_votes_group = election_data_df[["state","county","total_votes"]]
election_votes_group.head()                               
```

```python
state_votes_group = election_votes_group.groupby(["state"])
state_votes_sum = state_votes_group["total_votes"].sum()
state_votes_sum.head() 
```

```python
county_votes_group = election_votes_group.groupby(["state", "county"])
county_votes_sum = county_votes_group["total_votes"].sum()
county_votes_sum 
```

```python
biden_votes_group = election_data_df.loc[election_data_df["candidate"]=="Joe Biden", :]
biden_votes_group.head()
```

```python
biden_state_group = biden_votes_group.groupby(["state"])
biden_state_total = biden_state_group["total_votes"].sum()
biden_state_total
```

```python
biden_county_group = biden_votes_group.groupby(["state", "county"])
biden_county_total = biden_county_group["total_votes"].sum()
biden_county_total
```

```python
trump_votes_group = election_data_df.loc[election_data_df["candidate"]=="Donald Trump", :]
trump_votes_group.head()
```

```python
trump_state_group = trump_votes_group.groupby(["state"])
trump_state_total = trump_state_group["total_votes"].sum()
trump_state_total
```

```python
trump_county_group = trump_votes_group.groupby(["state", "county"])
trump_county_total = trump_county_group["total_votes"].sum()
trump_county_total
```

```python
# Creating a new DataFrame using both biden and trump state totals
state_votes_total = pd.DataFrame({"biden_state_votes": biden_state_total,
                                    "trump_state_votes": trump_state_total,
                                     "total_state_votes": state_votes_sum})
state_votes_total.head()
```

```python
state_votes_total.loc[state_votes_total['biden_state_votes'] > state_votes_total['trump_state_votes'],
                      'state_winner'] = 'biden'
state_votes_total.loc[state_votes_total['trump_state_votes'] > state_votes_total['biden_state_votes'],
                      'state_winner'] = 'trump'
state_votes_total['biden_state_percent']= state_votes_total['biden_state_votes']/state_votes_total['total_state_votes']
state_votes_total['trump_state_percent']= state_votes_total['trump_state_votes']/state_votes_total['total_state_votes']
state_votes_total.head()
```

```python
# Creating a new DataFrame using both biden and trump county totals
county_votes_total = pd.DataFrame({"biden_county_votes": biden_county_total,
                                    "trump_county_votes": trump_county_total,
                                    "total_county_votes": county_votes_sum})
county_votes_total.head()
```

```python
county_votes_total.loc[county_votes_total['biden_county_votes'] > county_votes_total['trump_county_votes'],
                      'county_winner'] = 'biden'
county_votes_total.loc[county_votes_total['trump_county_votes'] > county_votes_total['biden_county_votes'],
                      'county_winner'] = 'trump'
county_votes_total['biden_county_percent']= county_votes_total['biden_county_votes']/county_votes_total['total_county_votes']
county_votes_total['trump_county_percent']= county_votes_total['trump_county_votes']/county_votes_total['total_county_votes']
county_votes_final = county_votes_total
county_votes_final.head()
```
