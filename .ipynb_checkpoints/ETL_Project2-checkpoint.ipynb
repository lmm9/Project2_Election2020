```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np 
```

```python
#Loading csv containing all total number of votes form every state of the 2020 election data 
csv_file = "president_state.csv"
statevotemap = pd.read_csv(csv_file, encoding='ISO-8859-1')
statevotemap.head()
```

```python
#Esport index reset 
statevotemap = statevotemap.reset_index()
statevotemap.head()
```

```python
#Setting Index to its own column
statevotemap = statevotemap.rename(columns = {'index':'id'})
statevotemap.set_index('id', inplace = True)
statevotemap.head()
```
